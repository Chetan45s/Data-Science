```python
train_clean = train_clean.fillna(train_clean["Age"].mean())
```

```python
# Creating DataFrames for storing input and output data

input_cols = ["Pclass","Sex","Age","SibSp","Parch","Fare"]
output_cols = ["Survived"]

X = train_clean[input_cols]
Y = train_clean[output_cols]

print(X.shape,Y.shape)
print(type(X))
```

```python
def entropy_calculation(col):
        
    counts = np.unique(col,return_counts=True)
    N = float(col.shape[0])
    
    ent = 0.0
    
    for ix in counts[1]:
        p  = ix/N
        ent += (-1.0*p*np.log2(p))
    
    return ent
```

```python
# Train_clean is the input table before spliting
# col_name -> On the basis of which table we are spliting the data
# col_val  -> Dividing criteria, most of the time its mean of the col_name

def divide_data(train_clean,col_name,col_val):
    
    # creating two empty df
    x_right = pd.DataFrame([],columns=train_clean.columns)    
    x_left = pd.DataFrame([],columns=train_clean.columns)
    
    for itr in range(train_clean.shape[0]):
        
        val = train_clean[col_name].loc[itr]
        
        if val > col_val:
            x_right = x_right.append(train_clean.loc[itr])
        else:
            x_left = x_left.append(train_clean.loc[itr])
    return x_left,x_right
```

```python
divide_data(train_clean.head(10),"Sex",0.5)
```

```python
def information_gain(train_clean,col_name,col_val):
    
    x_left,x_right = divide_data(train_clean,col_name,col_val)
    
    l = float(x_left.shape[0]/train_clean.shape[0])    
    r = float(x_right.shape[0]/train_clean.shape[0])
    
    if x_left.shape[0] == 0 or x_right.shape[0] ==0:
        return -1000000 #Min Information Gain
    
    i_gain = entropy_calculation(train_clean.Survived) - (l*entropy_calculation(x_left.Survived)+r*entropy_calculation(x_right.Survived))
    return i_gain
    
```

```python
for fx in X.columns:
    print(fx)
    print(information_gain(train_clean,fx,train_clean[fx].mean()))
```

```python
class DecisionTree:
    
    #Constructor
    
    # So here is the case why we have restricted the maxx_depth upto 5 only.
    # while constructing a tree it can go large depth and causing overfitting of the data.
    # In order to prevent this we use methods :
    #        1) Define a max_Depth
    #       2) Contruct a tree and remove values from the end.
    
    # Here we are using the case 1
    
    def __init__(self,depth=0,max_depth=5):
        self.left = None
        self.right = None
        self.fkey = None
        self.fval = None
        self.max_depth = max_depth
        self.depth = depth
        self.target = None
        
    def train(self,X_train):
        
        features = ['Pclass','Sex','Age','SibSp', 'Parch', 'Fare']
        info_gains = []
        
        for ix in features:
            i_gain = information_gain(X_train,ix,X_train[ix].mean())
            info_gains.append(i_gain)
            
        self.fkey = features[np.argmax(info_gains)]
        self.fval = X_train[self.fkey].mean()
#         print("Making Tree Features is : ",self.fkey)
        
        #Split Data
        data_left,data_right = divide_data(X_train,self.fkey,self.fval)
        data_left = data_left.reset_index(drop=True)
        data_right = data_right.reset_index(drop=True)
         
        #Truly a left node
        if data_left.shape[0]  == 0 or data_right.shape[0] ==0:
            if X_train.Survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        #Stop earyly when depth >=max depth
        if(self.depth>=self.max_depth):
            if X_train.Survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        
        #Recursive Case
        self.left = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.left.train(data_left)
        
        self.right = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.right.train(data_right)
        
        #You can set the target at every node
        if X_train.Survived.mean() >= 0.5:
            self.target = "Survive"
        else:
            self.target = "Dead"
        return
    
    def predict(self,test):
        if test[self.fkey]>self.fval:
            #go to right
            if self.right is None:
                return self.target
            return self.right.predict(test)
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
```

```python
split = int(0.7*train_clean.shape[0])
train_data = train_clean[:split]
test_data = train_clean[split:]
test_data = test_data.reset_index(drop=True)
```

```python
print(train_data.shape,test_data.shape)
```

```python
dt = DecisionTree()
dt.train(train_data)
```

```python
print(dt.fkey)
print(dt.fval)
print(dt.left.fkey)
print(dt.right.fkey)
```

```python
y_pred = []
for ix in range(test_data.shape[0]):
    y_pred.append(dt.predict(test_data.loc[ix]))
y_pred
```
